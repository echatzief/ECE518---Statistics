### Using graphics in plotting

```r
x <- seq(-3, 3, by = 0.1)
y <- x
z <- outer(x, y, function(xi, yj) xi^2+yj^2)
persp(x, y, z)
        
y=x
f=outer(x,y,function(x,y)cos(y)/(1+x^2))
z <- outer(x, y, function(xi, yj) xi^2+yj^2)
persp(x, y, z)
contour(x,y,f)
contour(x,y,f,nlevels=45,add=T)
fa=(f-t(f))/2
contour(x,y,fa,nlevels=15)
           
image(x,y,fa)
persp(x,y,fa,theta=30,phi=40)
```
