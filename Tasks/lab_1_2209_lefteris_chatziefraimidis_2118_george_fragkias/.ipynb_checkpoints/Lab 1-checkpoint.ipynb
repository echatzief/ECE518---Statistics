### Loading Libraries

```r
library(ISLR)
library(dplyr)
library(readr)
library(ggplot2)
library(GGally)
library(manipulate)
```

### Loading Data

```r
auto = read.table("Auto.data")
auto = read.table("Auto.data",header=T,na.strings="?")
auto=read.csv("Auto.csv",header=T,na.strings="?")
dim(auto)
auto[1:4,]
```

### Removing NA values from the dataset

```r
auto=na.omit(auto)
auto
```

```r
dim(auto)
names(auto)
```

### Print subset of the data

```r
library(dplyr)
head(auto)
tail(auto)

# "The first row":
auto[1,]

# "All rows in second and fifth columns":
auto[,c(2,5)]

# "The first row in second to fifth columns":
auto[1,2:5]
```

### Use CSV and print data

```r
auto=read.csv("Auto.csv",header=T,na.strings="?")

head(auto)
tail(auto)

# "The first row":
auto[1, ]
# "All rows in second and fifth columns":
auto[, c(2, 5)]
# "The first row in second to fifth columns":
auto[1, 2:5]
```

### Using as.factor functionality

```r
library(sf)
library(dbplyr)
library(magrittr)
require(magrittr)
auto$origin = as.factor(auto$origin)
auto$origin
summary(auto)
```

### Calculating the mean values of a column

```r
mean(auto$displacement, data=auto)
```

### Plotting

```r
library(ggplot2)
ggplot(auto) + geom_point(aes(x=cylinders, y=mpg))
```

```r
auto$cylinders = as.factor(auto$cylinders)
auto$cylinders
```

### Using boxplots

```r
ggplot(auto) + geom_boxplot(aes(x=cylinders, y=mpg)) + xlab("Cylinders") + ylab("MPG")
```

### Histogram for mpg

```r
ggplot(auto) + geom_histogram(aes(x=mpg))
```

```r
ggplot(auto) + geom_histogram(aes(x=mpg),binwidth=5)
```

```r
library(GGally)
nncols = 9
auto = na.omit(auto)
ggpairs(auto,columns = 1:nncols,cardinality_threshold = 500)
```
